```python
# Лабораторная работа №1
#
# Выполнила: Гришина А.А.
# Группа: ЦИБ-251
# Задание №7.Текст задания: Стек. Симуляция текстового редактора с операциями ввода и Undo (отмена).

### Реализация

### Класс Stack
#- push(item) — добавить элемент — O(1)
#- pop() — удалить элемент — O(1)
#- is_empty() — проверка на пустоту — O(1)
#- clear() — очистка стека — O(1)

### Класс TextEditor
#- type_text(chars) — ввод текста — O(1)
#- undo() — отмена действия — O(1)
#- redo() — повтор действия — O(1)
#- get_text() — получить текст — O(1)
#- reset() — сброс редактора — O(1)

class Stack:
    def __init__(self):
        self._items = []

    def push(self, item):
        self._items.append(item)

    def pop(self):
        if self.is_empty():
            raise IndexError("Stack is empty")
        return self._items.pop()

    def is_empty(self):
        return len(self._items) == 0

    def clear(self):
        self._items.clear()
```

```python
class TextEditor:
    def __init__(self):
        self._current_text = ""
        self._undo_stack = Stack()
        self._redo_stack = Stack()

    def type_text(self, chars):
        if not chars:
            return
        self._undo_stack.push(self._current_text)
        self._current_text += chars
        self._redo_stack.clear()

    def undo(self):
        if self._undo_stack.is_empty():
            print("Нечего отменять!")
            return
        self._redo_stack.push(self._current_text)
        self._current_text = self._undo_stack.pop()

    def redo(self):
        if self._redo_stack.is_empty():
            print("Нечего повторять!")
            return
        self._undo_stack.push(self._current_text)
        self._current_text = self._redo_stack.pop()

    def get_text(self):
        return self._current_text

    def reset(self):
        self._current_text = ""
        self._undo_stack.clear()
        self._redo_stack.clear()
```

```python
def main():
    editor = TextEditor()
    print("=== Текстовый редактор с Undo/Redo ===")
    print("Команды:")
    print("  t <текст>   - ввод текста")
    print("  u           - Undo (отмена)")
    print("  r           - Redo (повтор)")
    print("  p           - показать текст")
    print("  reset       - сброс")
    print("  exit        - выход")

    while True:
        cmd = input("\n> ").strip()

        if cmd == "exit":
            print("Выход")
            break
        elif cmd == "u":
            editor.undo()
            print(f"Текст: '{editor.get_text()}'")
        elif cmd == "r":
            editor.redo()
            print(f"Текст: '{editor.get_text()}'")
        elif cmd == "p":
            print(f"Текст: '{editor.get_text()}'")   # <- ИСПРАВЛЕНО
        elif cmd == "reset":
            editor.reset()
            print("Редактор сброшен")
        elif cmd.startswith("t "):
            text = cmd[2:]
            editor.type_text(text)
            print(f"Текст: '{editor.get_text()}'")
        else:
            print("Неизвестная команда. Используйте: t текст, u, r, p, reset, exit")

main()
```
