```python
import hashlib

def is_hash_equal_case_insensitive(s1: str, s2: str) -> bool:
    # Вычисляем SHA1 хеши для исходных строк (с учетом регистра)
    hash1 = hashlib.sha1(s1.encode()).hexdigest()
    hash2 = hashlib.sha1(s2.encode()).hexdigest()

    # Сравниваем хеши
    return hash1 == hash2

# Тесты
print("Тест 1:", is_hash_equal_case_insensitive("Test", "test"))  # Ожидается: False
print("Тест 2:", is_hash_equal_case_insensitive("ABC", "ABC"))    # Ожидается: True

# Дополнительные тесты
print("\nДополнительные тесты:")
print("'hello' vs 'hello':", is_hash_equal_case_insensitive("hello", "hello"))     # True
print("'Hello' vs 'hello':", is_hash_equal_case_insensitive("Hello", "hello"))     # False
print("'Python' vs 'PYTHON':", is_hash_equal_case_insensitive("Python", "PYTHON")) # False
```
