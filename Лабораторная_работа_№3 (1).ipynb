```python
# Лабораторная работа №3
# Задание 7.
# Реализуйте функцию bubble_sort_desc_len(arr), которая сортирует список строк arr по убыванию их длины (от самых длинных к самым коротким).
# Выполнил: Гришина А.А.
# Группа: ЦИБ-251

def bubble_sort_desc_len(arr):
    """
    Сортирует список строк методом пузырька по убыванию длины строк.
    Алгоритм:
    - Проходит по массиву, сравнивая соседние элементы.
    - Если длина текущей строки меньше длины следующей, меняет их местами.
    - Повторяет проходы, пока массив не будет отсортирован.
    """
    n = len(arr)
    for i in range(n - 1):
        swapped = False
        for j in range(n - 1 - i):
            # Сравниваем длины строк (по убыванию)
            if len(arr[j]) < len(arr[j + 1]):
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swapped = True
        # Если за проход не было обменов — массив отсортирован
        if not swapped:
            break


if __name__ == "__main__":
    # Тест 1: из условия задачи
    a = ["a", "ccc", "bb"]
    bubble_sort_desc_len(a)
    print("Тест 1:", a)
    # Ожидается: ['ccc', 'bb', 'a']

    # Тест 2: пустой список
    b = []
    bubble_sort_desc_len(b)
    print("Тест 2:", b)
    # Ожидается: []

    # Тест 3: уже отсортированный по убыванию
    c = ["longest", "medium", "short"]
    bubble_sort_desc_len(c)
    print("Тест 3:", c)
    # Ожидается: ['longest', 'medium', 'short']

    # Тест 4: обратный порядок (по возрастанию длины)
    d = ["hi", "hello", "a", "world!"]
    bubble_sort_desc_len(d)
    print("Тест 4:", d)
    # Ожидается: ['world!', 'hello', 'hi', 'a']

    # Тест 5: строки с одинаковой длиной (не должны менять порядок)
    e = ["cat", "dog", "bat"]
    bubble_sort_desc_len(e)
    print("Тест 5:", e)
    # Ожидается: ['cat', 'dog', 'bat'] (т.к. длины равны)

    print("\nВсе тесты выполнены успешно!")
```
